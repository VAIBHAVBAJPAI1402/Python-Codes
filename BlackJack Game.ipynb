```python
#improrting random and sys module for generating random values from cards list
import random
import sys

#Game class represents player
class Game(object):
    #class object attributes
    player_name = ''
    player_bet = 0
    player_stack = 0
    
    #class methods
    def __init__(self,pname,pbet):
        Game.player_name = pname
        Game.player_bet = pbet
        print('Player initialized !')
    def update_stack(self,value):
        Game.player_stack = value
    def update_bet(self,amt):
        Game.player_bet += amt

#Dealer class represents Automated Dealer
class Dealer(object):
    #class object attributes
    dealer_name = 'Dealer'
    dealer_stack = 0
    #class methods
    def __init__(self):
        print('Dealer initialized !')
    def update_stack(self,value):
        Dealer.dealer_stack = value
#this function chooses 2 draw cards for player and also allows him to improve their hand   
def turn(player,ace,pip_cards,face_cards,add = 0):
    a1 = [ace,pip_cards,face_cards]
    if add==0:  
        #choosing and adding 2 cards to player's hand
        card_cat1 = random.choice(a1)
        card_cat2 = random.choice(a1)
        card1 = random.choice(card_cat1)
        card_cat1.remove(card1)
        card2 = random.choice(card_cat2)
        card_cat2.remove(card2)
        Game.player_stack += (card1 + card2)
        print('Player Card value =',Game.player_stack)
        if Game.player_stack>=21:
            print('-'*50)
            print('Player Card total exceeeded 21')
            print('PLAYER LOST')
            print('AMOUNT LOST =',Game.player_bet)
            sys.exit('Game Over')
    else:
        #adding one more card to improve the players hand
        add_cat = random.choice(a1)
        add_card = random.choice(add_cat)
        add_cat.remove(add_card)
        Game.player_stack += add_card
        if Game.player_stack>=21:
            #checking for hand bursts for player
            print('-'*50)
            print('Player Card value =',Game.player_stack)
            print('Player Card total exceeeded 21')
            print('PLAYER LOST')
            print('AMOUNT LOST =',Game.player_bet)
            sys.exit('Game Over')
        else:
            print('-'*50)
            print('Card added successfully!')
            print('Player hand improved')
            print('Player Card total =',Game.player_stack)
        
    ch = input('Do you want to update betting amount(y/n)')
    if ch=='y' or ch=='Y':
        #updating betting amount
        amt = int(input('Enter new betting amount = '))
        player.update_bet(amt)
        print('Bet Updated')
        print('New Amount =',Game.player_bet)
    else:
        print('OK no updation in betting amount')

#this function allows the dealer to pick cards from the deck until its stack value becomes 17 or more            
def dealer_turn(dealer,ace,pip_cards,face_cards):
    
    d1 = [ace,pip_cards,face_cards]
    while(Dealer.dealer_stack<17):
        card_cat = random.choice(d1)
        card_dealer = random.choice(card_cat)
        Dealer.dealer_stack += card_dealer
        card_cat.remove(card_dealer)
    print('Cards picked by dealer -')
    print('Dealer stack value = ',Dealer.dealer_stack)
    if Dealer.dealer_stack>=21 and Game.player_stack<21:
        #checking player stack value with dealer stack value
        print('-'*50)
        print('PLAYER WON')
        print('AMOUNT WON =',Game.player_bet)
        sys.exit('Game Over')
    
#start function          
def start():
    #initialising card stacks
    ace = [11]*4
    pip_cards = [2,3,4,5,6,7,8,9,10]*4
    face_cards = [10]*3
    #player name and betting amount
    name = input('Enter player name = ')
    bet = int(input('Enter betting amount = '))
    print('--'*50)
    player = Game(name,bet)
    turn(player,ace,pip_cards,face_cards)
    ch = input('Do you want to improve your hand(y/n) = ')
    if ch=='y' or ch=='Y':
        turn(player,ace,pip_cards,face_cards,1)
    if player.player_stack==21:
        #checking for hand burst
        print('Hand burst')
        sys.exit('GAME OVER - YOU LOSE')
    else:
        dealer = Dealer()
        dealer_turn(dealer,ace,pip_cards,face_cards)
            
    if(Dealer.dealer_stack < Game.player_stack):
        print('-'*50)
        print('PLAYER WON')
        print('AMOUNT WON =',Game.player_bet)
    elif(Dealer.dealer_stack == Game.player_stack):
        #draw condition
        print('-'*50)
        print(' IT\'s A DRAW')
    else:
        print('-'*50)
        print('YOU LOSE')
        print('AMOUNT LOST =',Game.player_bet)
        

start()
```
